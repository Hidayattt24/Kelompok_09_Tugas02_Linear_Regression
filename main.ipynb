# üõ†Ô∏è Eksplorasi Data dan Pra-pemrosesan

## üõ†Ô∏è **Pilih Kolom yang Relevan untuk Analisis**
Karena kita hanya tertarik pada target (Sales) dan fitur yang mempengaruhinya (Price, Discount, Rating, NumReviews), kita buat dataset baru yang hanya berisi kolom-kolom tersebut:

```python
# Pilih hanya kolom yang relevan untuk analisis
selected_columns = ["Sales", "Price", "Discount", "Rating", "NumReviews"]
df_clean = df[selected_columns]  # Menggunakan raw_data sebagai dataset awal

# Tampilkan 5 data pertama untuk verifikasi
print(df_clean.head())
```

## üõ†Ô∏è **Menangani Missing Values (Jika Ada)**  
Sebelum membangun model machine learning, penting untuk memastikan bahwa dataset tidak memiliki **missing values** (nilai yang hilang). Missing values dapat menyebabkan bias dalam analisis dan memengaruhi performa model prediktif.  

Untuk menangani missing values, kita dapat melakukan langkah-langkah berikut:  
1. **Menampilkan jumlah missing values di setiap kolom**  
2. **Menghapus atau mengisi nilai yang hilang berdasarkan strategi yang sesuai**  
   - Menghapus baris/kolom dengan terlalu banyak missing values  
   - Mengisi nilai yang hilang dengan **mean/median** (untuk numerik) atau **modus** (untuk kategorikal)  

Berikut adalah kode untuk mengecek dan menangani missing values:

```python
# Cek missing values dalam dataset
missing_values = df_clean.isnull().sum()
print("Missing values per kolom:\n", missing_values)
```

‚úÖ Penjelasan:

Tidak ada nilai missing di setiap kolom.

Jika ada nilai kosong, perlu dilakukan imputasi atau penghapusan data.

## üõ†Ô∏è **Menangani Outlier**  
Sebelum menghapus atau menangani outlier, kita perlu mengidentifikasi outlier menggunakan Interquartile Range (IQR):

```python
# Menghitung IQR
Q1 = df_clean.quantile(0.25)
Q3 = df_clean.quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menandai outlier
outliers = ((df_clean < lower_bound) | (df_clean > upper_bound))

# Menampilkan jumlah outlier per kolom
print("Jumlah outlier per kolom:\n", outliers.sum())
```

‚úÖ Penjelasan:

IQR digunakan untuk menentukan nilai yang dianggap outlier (di luar rentang normal).

Tidak ada outlier yang terdeteksi pada data ini.

## üõ†Ô∏è**Visualisasi Outlier dengan Boxplot**
Setelah mendeteksi outlier, kita bisa melihat distribusi datanya dengan boxplot:

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_clean)
plt.title("Boxplot untuk Deteksi Outlier")
plt.xticks(rotation=45)
plt.show()
```

‚úÖ Penjelasan:

Boxplot menunjukkan distribusi data untuk setiap kolom.

Tidak ada titik ekstrem yang muncul di luar whiskers, menandakan tidak adanya outlier.

## üõ†Ô∏è**Menghitung Persentase Outlier dengan IQR**
Untuk memahami dampak outlier dalam dataset:

```python
# Menghitung persentase outlier per kolom
outlier_percentage = (outliers.sum() / len(df_clean)) * 100
print("Persentase outlier per kolom:\n", outlier_percentage)
```

‚úÖ Penjelasan:

Persentase outlier di semua kolom adalah 0%, artinya semua data dalam rentang normal.

Jika ada outlier, bisa dilakukan penanganan seperti transformasi atau penghapusan data ekstrem.

## üõ†Ô∏è**Normalisasi atau Standarisasi Data (Jika Diperlukan)**
‚úÖ Kenapa dibutuhkan?

Normalisasi atau standarisasi diperlukan jika skala antar fitur sangat berbeda.

Contoh: Sales memiliki nilai ribuan, sementara Discount hanya dalam rentang 0-1.

üìå Kode (Standarisasi dengan StandardScaler)

```python
from sklearn.preprocessing import StandardScaler

# Standarisasi hanya untuk kolom numerik kecuali target (Sales)
scaler = StandardScaler()
df_clean_scaled = df_clean.copy()
df_clean_scaled[["Price", "Discount", "Rating", "NumReviews"]] = scaler.fit_transform(df_clean[["Price", "Discount", "Rating", "NumReviews"]])

# Tampilkan data setelah standarisasi
print(df_clean_scaled.head())
```

üìå Penjelasan:

Kolom Price, Discount, Rating, dan NumReviews telah distandarisasi (rata-rata 0, standar deviasi 1).

Sales tidak distandarisasi karena merupakan target prediksi.
